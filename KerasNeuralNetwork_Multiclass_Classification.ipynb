```python
# First neural network with keras tutorial
import pandas as pd 
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.vis_utils import plot_model
```

```python
df = pd.read_excel('D:/PGR210_Project/Dry_Bean_Dataset.xlsx')
```

```python
from sklearn.preprocessing import StandardScaler
from keras.utils.np_utils import to_categorical
x = df.drop("Class", axis = 1)
sc = StandardScaler()
x = pd.DataFrame(sc.fit_transform(x))
y = df['Class']
y_cat = to_categorical(y)
# Class column must have numbers instead of string
# These numbers must be converted to categorical data
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x.values, y_cat, test_size=0.2, random_state=5)
```

In the next part I needed to change three crucial parameters:
1- input_shape = it must be number of independent varibles
2- The last 'model.add' parameters (output). The number must be 'number of classes' + 1. 
Because there is already one 'class' column in the dataset.
3- As a loss I had to select "categorical_crossentropy".

```python
from tensorflow.keras.optimizers import Adam
model = Sequential()
model.add(Dense(60, input_shape = (16,), activation = "relu"))
model.add(Dense(15, activation = "relu"))
model.add(Dense(15, activation = "relu"))
model.add(Dense(8, activation = "softmax"))
model.compile(Adam(learning_rate = 0.01), "categorical_crossentropy", metrics = ["accuracy"])
#model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train, y_train, verbose=1, epochs=100, batch_size = 100)
```

```python
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', loss)
print('Test accuracy:', accuracy*100)
```
